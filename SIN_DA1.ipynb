```python
#DETECTION OF COMMUNITIES IN A FACEBOOK DATASET
import networkx as nx
import csv
import matplotlib.pyplot as plt
import community

G=nx.Graph()
with open('F:\\VIT\\FALL SEM 20-21\\SIN\\THEORY\\DA\\facebook_combined.txt','r') as csvfile:
    plots=csv.reader(csvfile, delimiter=' ')
    for row in plots:
        G.add_edge(row[0],row[1])

print(nx.info(G))
pos=nx.spring_layout(G) 
plt.figure(figsize=(10, 10))
plt.axis('off') 
nx.draw_networkx(G,pos,with_labels=False,node_size=20) 
parts=community.best_partition(G)
pos=nx.spring_layout(G)
plt.figure(figsize=(10, 10))
plt.axis('off')
nx.draw_networkx_nodes(G,pos,node_size=20,cmap=plt.cm.gist_rainbow,node_color=list(parts.values()))
nx.draw_networkx_edges(G,pos,alpha=0.3)
plt.show(G)
```
